```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
#import geopandas as gpd
import os
import re
from matplotlib.ticker import MaxNLocator
from load_files import get_measurements
path = "C:/Users/20175825/OneDrive - TU Eindhoven/Data Science Bachelor Year 3/Data Challenge 3/waterschap-aa-en-maas_sewage_2019/"

%matplotlib inline
```

```python
def get_rain(path, convert_time=False):
    files = os.listdir(path)
    data = [pd.read_csv(path + "/" + i, sep = ",", skiprows=2) for i in files]
    data =  pd.concat(data, axis= 0, join= 'outer')

    if convert_time == True:
        data["Begin"] = data.apply(lambda row: datetime.datetime.strptime(row["Begin"], '%d-%m-%Y %H:%M:%S'), axis = 1)
        data["Eind"] = data.apply(lambda row: datetime.datetime.strptime(row["Eind"], '%d-%m-%Y %H:%M:%S'), axis = 1)
        data.set_index('Begin', inplace= True)
        data.sort_index(inplace=True)
    return data

rain = get_rain(path + "sewer_data/rain_timeseries", True)
rain.groupby([rain.index.year, rain.index.month, rain.index.day, rain.index.hour]).sum()
```

# Streetnames -> longitude Latitude

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="Tom")
location= geolocator.geocode("Lekerstraat")
print(location.address)
print((location.latitude, location.longitude))
```

```python
lst = rain.columns.str.replace('\d+', '').str.replace('.', '').str.replace('+', '').str.strip()[3:]##.str.replace('', 'Nederland')
ned_lst = [i + ', The Netherlands' for i in lst]
```

```python


#list(rain.columns)[3:]
streetnames = list(rain.columns.str.replace('\d+', '').str.replace('.', '').str.replace('+', '').str.strip())[3:]
exact_loc = []
longitude = []
latitude = []
count = 0 
for street in ned_lst:
    print(count/len(streetnames))
    try:
        location = geolocator.geocode(street)
        exact_loc.append(location.address)
        longitude.append(location.longitude)
        latitude.append(location.latitude)
    except: 
        exact_loc.append("NAN")
        longitude.append("NAN")
        latitude.append("NAN")
    count += 1


```

```python
df_streets = pd.DataFrame()
df_streets['Streetname'] = ned_lst
df_streets['Exact Location'] = exact_loc
df_streets['Longitude'] = longitude
df_streets['Latitude'] = latitude

df_streets
```
